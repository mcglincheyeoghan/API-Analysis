```python
'''
import json and requests packages so that
dataset can be saved in JSON format and 
to allow data from API to be collected
'''
import json
import requests
```

```python
'''
send a request and get a response object,
gets current weather data (27th October 2020)
from 11 cities: Dublin City, East London, Paris,
Berlin Tempelhof, San Francisco, Canberra, Beijing,
Pyongyang, Rio de Janeiro, Cairo and Mexico City
'''
response = requests.get('http://api.openweathermap.org/data/2.5/group?id=7778677,1006984,6618607,7290253,5391959,2172517,1816670,1871856,3451189,360630,3530597&appid=3019730f33cf5f61ca167aba43b629ff')
```

```python
#determine if response was successful by getting status code, successful if integer returned
response.status_code
```

```python
#get data in a more readable JSON format
response.json()
```

```python
'''
save the collected dataset as a JSON file
by using a with statement to open a destination
file called 'dataset.json', then use json.dump
to write the data object to the outfile file
'''
with open('dataset.json', 'w') as outfile:
    json.dump(response.json(), outfile)
```

```python
#import relevant packages
import json
import pandas as pd
import time
import numpy as np

#read the data as a JSON file
df = pd.read_json("dataset.json")

'''
create Data Frame from the data that has been read
by taking the 'list' key and all of its values and
converting it to a list, then use the from_records
funciton to convert the Data Frame to a Data Frame 
object called 'data'
'''
data = pd.DataFrame(df.list.values.tolist())
data = pd.DataFrame.from_records(data)
data
```
